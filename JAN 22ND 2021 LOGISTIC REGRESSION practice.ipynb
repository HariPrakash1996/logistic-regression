# LOGISTIC REGRESSION

# def:


* it is one of the most popular machine learning technique under supervised learning algorithm
* it is used to solve the classification problem
* logistic regression produces results in binary format 
* which is used to predict the outcome of a categorical dependent variable
* outcome must be discrete or categorical
* LOGISTIC REGRESSION GIVES THE PROBABILITY VALUES WHICH LIES BETWEEN  0 and 1
* it is a s_shaped logistic function
* here s_shape is called sigmoid function
* it is a mathematical function which is used to map the predicted values to probabilities
* it maps all the predictions probabilities values between 0 and 1
* logistic regression predicts the probabilities
* this algorithm is also called fisher scoring algorithm
#  THE VALUE OF LOGISTIC REGRESSION MUST BE 0 TO 1
* if all your coefficient values are 0 in sigmoid function 
* 0.5 is your threshold value

* yes or no
* 1 or 0

* humid or not humid

# ASSUMPTIONS:

* dependent variable must be categorical in nature
* the independent variable should not have any multi_collinearity

# types of logistic regression:

* Based on categories logistic regression can be classified into 3 types 
* they are
    * 1.BINOMIAL LOG_REG
    * def: in binomial log_reg two possible types of the dependent variables it contains they are : yes or no and true or false etc..,
    * 2. MULTINOMIAL LOG_REG
    * DEF: it has 3 or more possible unordered types of the dependent variable
    * ex: cat,dog,snake
    * 3. ORDINAL LOG_REG
    * DEF : it has 3 or more possible ordered types of the dependent variable
    * ex: low, medium,high

# difference between linear and ogistic regression

# linear regression:
* value of y is continuous variable
* variables we need to predict is continuous in nature
* it solves regression problems
* it is trying to predict the relationship between independent and dependent variable
* here we need to calculate the value of y with the value of x
* it has straight line graph


# logistic regression:

* value of y is categorical variables
* prediction should be discrete in nature
* it solves classification problems
* it has s_shaped curved
* here s_shape called sigmoid curve

```python
hd.dropna(inplace=True)
```

```python
x=hd.drop("Outcome",axis=True)
```

```python
y=hd["Outcome"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.get_params()
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report

classification_report(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
